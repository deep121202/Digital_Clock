```python

from tkinter import*


#import time library to obtain current time.
import time
import datetime

#create a function timing and variable current_time.

def timing():
    #display current hour, minute, seconds 
    current_time = time.strftime("%I:%M:%S %p")
    #configure the clock.
    clock.config(text=current_time)
    #clock will change after every 200 microseconds.
    clock.after (200, timing)

#Create a variable that will store our tkinter window.

win=Tk()
win.geometry("730x200") win.configure(background="black")

#create a variable 'clock' to show time.

clock=Label (win,font=("times 60 bold"),bg="black",fg="green")
clock.grid(row=2,column=2,ipadx=100)
timing()

#create a variable currenttime according to which greeting will be shown.

currenttime=datetime.datetime.now()

if currenttime.hour< 12:
    t9="Good morning..!"

elif 12 <= currenttime.hour < 18:
    t9="good afternoon..!"

else:
    t9="good evening..!"

greet=Label (win, text=t9, font="arial_24 ",bg="black",fg="green")
greet.grid(row=3, column=2)

win.mainloop()
```
